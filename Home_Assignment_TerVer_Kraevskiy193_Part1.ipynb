# Задание 1 

Пункт 1

```python
np.random.seed(123)                           
n = 100                                                
df = 5                                                  
x = chi2.rvs(size = n,                                  
             df = df)                                       
                                             

F_x = chi2.cdf(x, df = df)                              
plt.xlabel('x')                                         
plt.ylabel('F(x)')  
plt.title('Хи-квадрат выборочное')
plt.plot(x, F_x, '--', markersize = 1,                  
  label = "CDF", color = "orange")

```

```python
sns.histplot(x, stat = 'density',                   
                 color = "palevioletred",
                 label = "Хи-квадрат Гистограмма",
                 bins = 10)                             
plt.legend()
```

```python
np.random.seed(123)                      
n_st = 100                                              
df_st = 10                                                  
x_st = scipy.stats.t.rvs(size = n_st,                                  
             df = df_st)                                       
                                             

F_x_st = scipy.stats.t.cdf(x_st, df = df_st)                              
plt.xlabel('x')                                         
plt.ylabel('F(x)')  
plt.title('Cтюдент выборочное')
plt.plot(x_st, F_x_st, '--', markersize = 1,                  
  label = "CDF", color = "black")
```

```python
sns.histplot(x_st, stat = 'density',                   
                 color = "darkblue",
                 label = "Стьюдент Гистограмма  ",
                 bins = 10)                             
plt.legend()
```

```python
np.random.seed(123)                      
n_norm = 1000                                              
                                                
x_norm = norm.rvs(size = n_norm, loc = 10, scale = 25)                                       
                                             

F_x_norm = norm.cdf(x_norm, loc = 10, scale = 25 )                              
plt.xlabel('x')                                         
plt.ylabel('F(x)')  
plt.title('Нормальное выборочное')
plt.plot(x_norm, F_x_norm, '--', markersize = 1,                  
  label = "CDF", color = "limegreen")
```

```python
sns.histplot(x_norm, stat = 'density',                   
                 color = "red",
                 label = "Нормальное Гистограмма ",
                 bins = 10)                             
plt.legend()
```
